```python
#混合正規分布
from numpy.random import *
from matplotlib import pyplot as plt
import scipy as sp
import seaborn as sns
import numpy as np
%matplotlib inline
mu1 = [0, 0]
sigma1 = [[3, 2], [2, 1]]
mu2 = [5, 1]
sigma2 = [[2,3], [3, 2]]
r1 = 0.3
r2 = 1 - r1

# 2次元正規乱数を1万個生成
n = 10000
num1 = np.random.binomial(n, r1, 1)
num2 = n - num1
values1 = multivariate_normal(mu1, sigma1, num1)
values2 = multivariate_normal(mu2, sigma2, num2)
data = np.concatenate((values1,values2),axis=0)

#描画
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(values1[:,0],values1[:,1], c='red')
ax.scatter(values2[:,0],values2[:,1], c='blue')
ax.set_title('second scatter plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
fig.show()
```

```python
values1_x = values1[:,0]
values1_y = values1[:,1]
values1_z = values1_x + 1j * values1_y

values2_x = values2[:,0]
values2_y = values2[:,1]
values2_z = values2_x + 1j * values2_y
```

```python
values1_z_ = 2/(values1_z /2 +2j)
values2_z_ = 2/(values2_z /2 +2j)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(values1_z_.real,values1_z_.imag, c='red',alpha=0.5)
ax.scatter(values2_z_.real,values2_z_.imag, c='blue',alpha=0.5)
```
